```python
i = 1
while i <= 10:
  print(f'Valor de i en el ciclo {i}, {i}')
  i = i + 1
print(f'valor final de i es {i}')
```

```python

condicion = True
maximo = 0

while condicion:
  numero = int(input('ingresa un número: '))
  if numero > maximo:
    maximo = numero
  if numero == 0:
    print(f'El número mayor es, {maximo}')
    condicion = False
```

```python
numero =  int(input('ingresa un número: '))
cuenta = 0
suma = 0

while numero != 0:
  cuenta = cuenta+1
  suma = suma+numero
  numero =  int(input('ingresa un número: '))


print(f'FIN : Ingresaste {numero}')
print(f'RESULTADOS : Cantidad de números ingresados: {cuenta}, suma de número ingresados: {suma}')

```

```python

# range (start, stop, step)

# version 1, el start 0 y step 1
data = range(10) # range(0,10,1)

print(list(data))


# version 2, el start 3 y step 1
data = range(3,10) # range(3,10,1)

print(list(data))


# version 2, el start 3 y stop 10,  step 1
data = range(3,20,3) # range(3,10,1)

print(list(data))
```

```python
# Secuencias de números enteros.
data = range(20,-1,-2) # range(3,10,1)
print(list(data))
```

```python
numero =  int(input('ingresa un número: '))
coleccion = range(1,numero+1,1)

for i in coleccion:
  if numero%i == 0:
    print(f' Divisores: {i}')
```

```python
numero =  int(input('ingresa un número: '))
coleccion = range(1,numero+1,1)

for i in coleccion:
  salida = i / numero
  #print(f' Fraccionarios {i} / {numero}: {salida:.5f}')
  print(f' Fraccionarios {i} / {numero}: {round(salida,5)}')
```

# **Ciclos Anidados.**
Importante identanción


```python
nf = int(input('ingresa un número de filas: '))

f = 1

while f<=nf:
  c = 1
  while c<=f:
    print(c, end=' ')
    c = c + 1
  f = f + 1
  print()
```
